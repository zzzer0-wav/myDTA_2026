### Завдання 10. Перевірка простого числа ⭐⭐

Створіть функцію `is_prime(n)`, яка повертає `True`, якщо число просте, і `False` інакше.

**Нагадування:** просте число ділиться лише на 1 і на саме себе. Числа менші за 2 — не прості.

**Перевірте:**
```python
print(is_prime(7))    # True
print(is_prime(10))   # False
print(is_prime(1))    # False
```

> 💡 **Підказка:** У циклі від 2 до n-1 перевіряйте, чи ділиться `n` на якесь число без остачі. Якщо знайшли дільник — повертайте одразу `False`. Якщо цикл завершився без знайдених дільників — `True`.

````python
# Ваш код для завдання 10:
def is_prime(n):
    if n < 2:
        return False
    for num in range(2, n-1):
        if n % num == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
print(is_prime(1))
````

### Завдання 11. Факторіал ⭐⭐

Створіть функцію `factorial(n)`, яка обчислює факторіал числа `n`.

**Нагадування:** `n! = 1 * 2 * 3 * ... * n`. Наприклад, `5! = 120`.

**Перевірте:**
```python

print(factorial(0))   # 1
```

````python
# Ваш код для завдання 11:
def factorial(n):
    fact = 1
    for nums in range(1, n+1):
        fact *= nums
    return fact

print(factorial(5))
````

### Завдання 16. Найдовше слово ⭐⭐⭐

Створіть функцію `longest_word(sentence)`, яка приймає рядок-речення і повертає найдовше слово.

**Перевірте:**
```python
print(longest_word("Python це чудова мова програмування"))   # "програмування"
```

> 💡 **Підказка:** Розбийте речення на список слів через `.split()`. Потім переберіть і знайдіть найдовше, порівнюючи через `len()`.

````python
# Ваш код для завдання 16:
def longest_word(sentance):
    words = sentance.split()
    longest_word = ''
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word
    return longest_word

print(longest_word('asddas asdf aaaa a dsfdsfsdfsdfdsf'))
````

### Завдання 17. Калькулятор з операцією ⭐⭐⭐

Створіть функцію `calculate(a, b, operation)`, яка виконує одну з операцій залежно від рядка `operation`:
- `"+"` — додавання
- `"-"` — віднімання
- `"*"` — множення
- `"/"` — ділення

Якщо операція невідома — поверніть рядок `"Невідома операція"`. Якщо ділення на нуль — поверніть `"Помилка: ділення на нуль"`.

````python
# Ваш код для завдання 17:
def calculate(a, b, operation):

    if operation == '/' and b == 0:
        return 'Помилка: ділення на нуль!'
    elif operation == '+':
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '/':
        return a / b
    elif operation == '*':
        return a * b
    else:
        return 'Невідома операція'

print(calculate(2, 5, '/'))

````

### Завдання 18. Числа Фібоначчі ⭐⭐⭐

Створіть функцію `fibonacci(n)`, яка повертає **список** з перших `n` чисел Фібоначчі.

**Нагадування:** кожне число — сума двох попередніх. Перші два: 0 і 1.

**Перевірте:**
```python
print(fibonacci(7))   # [0, 1, 1, 2, 3, 5, 8]
```

> 💡 **Підказка:** Створіть список з перших двох чисел `[0, 1]`. У циклі додавайте суму двох останніх елементів через `.append()`.

````python
# Ваш код для завдання 18:
def fibonacci(n):
    if n <= 0:
        return []
    if n == 1:
        return [0]
    if n == 2:
        return [0, 1]

    fib = [0, 1]

    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])

    return fib

print(fibonacci(3))
````

### Завдання 22. Клас Circle ⭐

Створіть клас `Circle` з атрибутом `radius`. Додайте методи:
1. `area()` — площа кола (`π * r²`, використайте `3.14` або `math.pi`)
2. `circumference()` — довжина кола (`2 * π * r`)

````python
# Ваш код для завдання 22:
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return (math.pi * self.radius ** 2)

    def circumference(self):
        return (2 * math.pi * self.radius)

circle = Circle(5)

print(circle.area())
print(circle.circumference())
````

2. `average()` — повертає середній бал (0, якщо оцінок ще немає)
3. `__str__()` — повертає рядок виду `"Студент <name>, середній бал: <average>"`

Створіть студента, додайте 4 оцінки, виведіть об'єкт через `print()`.

````python
# Ваш код для завдання 23:
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def average(self):
        if len(self.grades) != 0:
            return sum(self.grades) / len(self.grades)
        else:
            return 0

    def __str__(self):
        return f"Студент {self.name}, середній бал: {self.average():.2f}"

st1 = Student('Ivan')

st1.add_grade(10)
st1.add_grade(8)
st1.add_grade(12)
st1.add_grade(8)

st2= Student('Maria')

print(st1)
print(st2)
````

### Завдання 25. Клас BankAccount ⭐⭐

Створіть клас `BankAccount` з атрибутами `owner` (ім'я власника) і `balance` (баланс, за замовчуванням 0). Додайте методи:
1. `deposit(amount)` — поповнити рахунок
2. `withdraw(amount)` — зняти кошти (якщо вистачає; інакше вивести повідомлення)
3. `show_balance()` — вивести `"Баланс <owner>: <balance> грн"`

Зробіть три операції і перевірте, що все працює.

> 💡 **Підказка:** У методі `withdraw` перевірте через `if amount > self.balance:` — якщо так, виведіть повідомлення і вийдіть з методу через `return`.

````python
# Ваш код для завдання 25:
class BankAccount():
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def show_balance(self):
        return f'Баланс {self.owner}: {self.balance} грн'

client1 = BankAccount('Maria', 200)

client1.deposit(100)
print(client1.show_balance())
````

2. `champion(teams)` — повертає команду з найбільшою кількістю очок
3. `total_matches(teams)` — загальна кількість зіграних матчів (сума всіх перемог + нічиїх + поразок, поділена на 2, бо кожен матч рахується двічі)
