# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

````python
rices = {
    "хліб": 25,
    "молоко": 38,
    "сир": 120,
    "масло": 95,
    "м'ясо": 280,
    "риба": 250,
    "яйця": 65
}
# Ваш код:
max_name = ""
max_price = 0

for product, price in prices.items():
  if max_price < price:
    max_price = price
    max_name = product

print(f'Найдорожчий товар - {max_name}, коштує {max_price}')
````

### Завдання 16. Підрахунок слів у тексті ⭐⭐

Дано рядок з текстом. Створіть словник, де ключ — слово, а значення — скільки разів воно зустрічається у тексті.

> 💡 **Підказка:** Використайте метод `.split()` — він розбиває рядок на список слів.

````python
text = "кіт пес кіт хом'як кіт пес папуга кіт пес рибка"
# Ваш код:
words = text.split()

word_count = {}

for word in words:
  if word in word_count.keys():
    word_count[word] += 1
  else:
    word_count[word] = 1
print(word_count)
````
