```python
# Запусти цей блок, щоб отримати датасет
import pandas as pd
import numpy as np

np.random.seed(42)
n = 300

categories = ['Молочне', 'Хліб та випічка', 'М\'ясо та риба', 'Овочі та фрукти', 'Напої']
products = {
    'Молочне':          ['Молоко 1л', 'Кефір 0.5л', 'Сир кисломолочний', 'Масло вершкове', 'Йогурт'],
    'Хліб та випічка':  ['Хліб білий', 'Батон', 'Булочка', 'Хліб житній', 'Круасан'],
    'М\'ясо та риба':   ['Куряче філе', 'Свинина', 'Ковбаса', 'Риба хек', 'Сосиски'],
    'Овочі та фрукти':  ['Картопля', 'Морква', 'Яблука', 'Банани', 'Помідори'],
    'Напої':            ['Вода 1.5л', 'Сік апельсиновий', 'Чай', 'Кава мелена', 'Лимонад'],
}
prices = {
    'Молоко 1л': 38, 'Кефір 0.5л': 22, 'Сир кисломолочний': 55, 'Масло вершкове': 95, 'Йогурт': 30,
    'Хліб білий': 28, 'Батон': 25, 'Булочка': 12, 'Хліб житній': 32, 'Круасан': 18,
    'Куряче філе': 165, 'Свинина': 210, 'Ковбаса': 145, 'Риба хек': 120, 'Сосиски': 98,
    'Картопля': 20, 'Морква': 15, 'Яблука': 45, 'Банани': 52, 'Помідори': 60,
    'Вода 1.5л': 18, 'Сік апельсиновий': 68, 'Чай': 85, 'Кава мелена': 155, 'Лимонад': 35,
}

chosen_cats = np.random.choice(categories, n, p=[0.25, 0.20, 0.20, 0.20, 0.15])
chosen_products = [np.random.choice(products[c]) for c in chosen_cats]

df_origin = pd.DataFrame({
    'transaction_id': range(1001, 1001 + n),
    'date':           pd.to_datetime(
                          np.random.choice(pd.date_range('2024-03-01', '2024-03-31'), n)
                      ),
    'hour':           np.random.choice(range(8, 21), n, p=[0.04,0.06,0.08,0.10,0.12,0.12,0.11,0.10,0.09,0.08,0.06,0.04,0.00]),
    'category':       chosen_cats,
    'product':        chosen_products,
    'quantity':       np.random.randint(1, 6, n),
    'price_uah':      [prices[p] for p in chosen_products],
    'customer_age':   np.random.randint(18, 70, n),
    'payment':        np.random.choice(['Готівка', 'Картка', 'Телефон'], n, p=[0.35, 0.50, 0.15]),
})
```

### Блок 4 - Нові стовпці та трансформація

Додай стовпець revenue_share - частка виторгу кожної транзакції від загального виторгу (у відсотках).

```python
total_revenue = df.total_uah.sum()
df['revenue_share'] = (df.total_uah / total_revenue * 100).round(4)

df[['product', 'total_uah', 'revenue_share']].head()
```

Додай стовпець age_group:

    'Молодь' - до 30 років
    'Середній вік' - 30–50 років
    'Старші' - понад 50 років


```python
df['age_group'] = pd.cut(
    df['customer_age'],
    bins=[0, 30, 50, 100],
    labels=['Молодь', 'Cередній вік', 'Старші'],
    right=False # [0, 30), [30, 50), [50, 100)
)

print(f'Розподіл по віковим групам:\n')
# df.head()
df.age_group.value_counts()
```

Додай стовпець weekday - день тижня з колонки date.

```python
df['weekday'] = df.date.dt.day_name()
print('Транзакція по дням тижня:\n')
df.weekday.value_counts()
```

Додай булевий стовпець is_big_purchase - True, якщо total_uah > 300.

```python
df['is_big_purchase'] = df.total_uah > 300
pct = df['is_big_purchase'].mean() * 100
print(f'Частка \'великих\' покупок (> 300) {pct:.1f}%')
```

```python
# Розподіл по віковим групам
print(f'Середній чек по віковим групам: ')
df.groupby('age_group', observed=True)['total_uah'].mean().round(2)
```

### Блок 6 - Фінальні висновки

Яка категорія товарів найприбутковіша?
___
Найприбутковіша категорія - "М'ясо та риба" з виручкою 25.223 грн, що становить майже 45% усіх продаж. Це пояснюється високими цінами за одиницю (98 - 210 грн).
Найменш прибутковіша категорія - "хліб та випічка" (4134 грн) оскільки це дешеві та щоденні товари.

Дослідити реальний прибуток неможливо, через відсутню собівартість товару, тому порівнюємо лише виручку.

```python
revenue_by_cat = df.groupby('category')['total_uah'].sum().sort_values(ascending=False)

print('Категорії по total_uah',revenue_by_cat)

revenue_by_cat_pct = (revenue_by_cat / revenue_by_cat.sum() * 100).round(1)

print('\nКатегорії по total_uah у відсотках',revenue_by_cat_pct)
```

Який спосіб оплати переважає?
___
За кількістю транзакцій переважає оплата карткою, більше половини клієнтів оплачусть саме так (52.3%), готівка на другому місті (34.7%) та телефоном оплачують менш за все (13%).

Проте середній чек майже однаковий для всіх методів - спосіб оплати у даному випадку не впливає на суму покупки.

```python
print('Сума чеку по оплаті:\n')
display(df.groupby('payment')['total_uah'].sum().round(2).sort_values(ascending=False))

print('\n\nСередній чек по оплаті:\n')
display(df.groupby('payment')['total_uah'].mean().round(2).sort_values(ascending=False))

print('\n\nСпіввідношення методів оплати у відсотках: \n')
display((df['payment'].value_counts(normalize=True) * 100).round(1).rename('частка_%'))
```
