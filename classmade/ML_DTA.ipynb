```
import numpy as np
import pandas as pd

data = {
    "Days": [1, 2, 3, 4, 5, 6, 7],
    "Coins": [0, 1, 0, 1, 100, 2, 3],
}
df = pd.DataFrame(data)

df
```

```
import matplotlib.pyplot as plt

plt.scatter(df['Days'], df['Coins'], color='Blue')
plt.title('Графік залежності')
plt.xlabel('Days')
plt.ylabel('Coins')
plt.show()
```

```
from sklearn.linear_model import LinearRegression

X = df[['Days']]
Y = df['Coins']

model = LinearRegression()
model.fit(X, Y)

# Вивід коефіціентів
print('Baгa (W):', model.coef_)
print('Зміщення (b):', model.intercept_)
```

```
# Прогноз 1.1
Y_pred = model.predict(X)

plt.scatter(X, Y, color='Blue')
plt.plot(X, Y_pred, color='Red')
plt.title('Лінійна регресія')
plt.xlabel('Days')
plt.ylabel('Coins')
plt.show()
```

```
# Прогноз 1.2
new_x =  [[10]]
predict_coins = model.predict(new_x)
print(f'Прогноз на {new_x[0][0]} день: {predict_coins[0]} монеток')
```
