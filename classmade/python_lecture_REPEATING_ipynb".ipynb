## 1.4. Цикл `for` — повторення дій

Цикл `for` дозволяє повторити блок коду декілька разів. Він дуже зручний, коли треба пройтись по чомусь — числах, літерах у слові, елементах списку.

### `for` може перебирати РІЗНІ речі!

Це важливо зрозуміти: `for` — універсальний інструмент, який вміє перебирати:
- 🔢 числові послідовності (через `range()`)
- 📝 текстові рядки (літера за літерою)
- 📋 списки (елемент за елементом — побачимо далі!)
- 📚 словники (ключ за ключем — теж побачимо!)

### 1.4.1. Перебір чисел через `range()`

Функція `range()` створює послідовність чисел:
- `range(5)` → 0, 1, 2, 3, 4 (від 0 до 5, але БЕЗ 5)
- `range(1, 6)` → 1, 2, 3, 4, 5 (від 1 до 6, без 6)
- `range(0, 10, 2)` → 0, 2, 4, 6, 8 (з кроком 2)

````python
# Найпростіший приклад - повторити 5 разів
for i in range(5):
    print("Крок номер", i)
````

````python
# Числа від 1 до 10
for number in range(1, 11):
    print(number)
````

````python
# Парні числа від 2 до 20
for n in range(2, 21, 2):
    print(n)
````

````python
# Сума чисел від 1 до 100
total = 0
for i in range(1, 101):
    total = total + i

print("Сума чисел від 1 до 100 =", total)
````

# Частина 2. Цикл `while` — нова тема! 🆕

## Чим `while` відрізняється від `for`?

- **`for`** використовуємо, коли ЗАЗДАЛЕГІДЬ ВІДОМО, скільки разів треба повторити дію (наприклад, 10 разів, або для кожної літери в слові).
- **`while`** використовуємо, коли треба повторювати дію ДОКИ ВИКОНУЄТЬСЯ ПЕВНА УМОВА.

### Аналогія з життя

> 🎯 **`for`:** «Зроби 20 присідань» — точна кількість.  
> 🎯 **`while`:** «Присідай, ПОКИ не втомишся» — невідомо скільки разів, залежить від умови.

### Синтаксис

```python
while умова:
    блок коду, який повторюється
```

⚠️ **ОБЕРЕЖНО:** Якщо умова ніколи не стане хибною — цикл буде виконуватись вічно (це називається «нескінченний цикл»). Завжди треба, щоб всередині циклу щось ЗМІНЮВАЛО умову!

````python
# Простий приклад: рахуємо від 1 до 5
counter = 1

while counter <= 5:
    print("Лічильник:", counter)
    counter = counter + 1   # ⚠️ Це обов'язково! Інакше нескінченний цикл

print("Цикл завершився!")
````

````python
# Зворотній відлік
seconds = 5

while seconds > 0:
    print(seconds, "...")
    seconds = seconds - 1

print("🚀 Старт!")
````

````python
# Приклад: користувач вводить пароль, поки не введе правильний
correct_password = "python123"
entered_password = ""
counter = 1

# Поки введений пароль НЕ дорівнює правильному - запитуємо знову
while entered_password != correct_password and counter <= 4:
    entered_password = input("Введіть пароль: ")
    if entered_password != correct_password:
        print("❌ Неправильно, спробуйте ще раз")
    counter = counter + 1

entered_password = input('Введіть пароль: ')

if entered_password == correct_password:
    print("✅ Доступ дозволено!")
else:
    print("❌ Неправильний пароль. Вас заблоковано на 1 годину. Спробуйте ввести пароль пізніше")
````

````python
# Приклад: подвоюємо число, поки воно менше 1000
number = 1

while number < 1000:
    print(number)
    number = number * 2

print("Число", number, "перевищило 1000, цикл зупинено")
````

### Корисні команди всередині циклів: `break` і `continue`

- **`break`** — негайно вийти з циклу (припинити виконання)
- **`continue`** — пропустити решту цього кроку і перейти до наступного

Ці команди працюють і з `while`, і з `for`.

````python
# break: виходимо з циклу, коли знайшли потрібне число
for number in range(1, 100):
    if number == 7:
        print("Знайшли 7! Виходимо з циклу")
        break
    print("Перевіряємо число", number)
````

````python
# continue: пропускаємо парні числа
for number in range(1, 11):
    if number % 2 == 0:
        continue   # пропускаємо парні
    print(number, "- непарне")
````

# Частина 3. Списки (`list`) — нова тема! 🆕

## Що таке список?

До цього ми зберігали по одному значенню в одній змінній. А якщо треба зберегти 100 імен студентів? Створювати 100 змінних? Звісно ні!

**Список** — це впорядкована колекція значень в одній змінній. Уявіть полицю в магазині: на ній стоїть багато товарів, кожен має своє місце (номер).

### Як виглядає список

Список створюється за допомогою квадратних дужок `[ ]`, елементи розділяються комами:

```python
fruits = ["яблуко", "банан", "апельсин"]
numbers = [10, 25, 3, 47, 8]
mixed = ["Олена", 25, True, 1.75]   # можна змішувати типи!
empty = []   # порожній список
```

### Індекси — номери елементів

Кожен елемент списку має свій НОМЕР (індекс). **Дуже важливо:** в Python нумерація починається з **0**, а не з 1!

```
fruits = ["яблуко", "банан", "апельсин"]
            0          1          2
```

````python
# Створюємо список фруктів
fruits = ["яблуко", "банан", "апельсин", "груша"]

# Виводимо весь список
print(fruits)

# Виводимо конкретні елементи за індексом
print("Перший фрукт:", fruits[0])      # яблуко
print("Другий фрукт:", fruits[1])      # банан
print("Третій фрукт:", fruits[2])      # апельсин
````

````python
# Від'ємні індекси - відлік з кінця
fruits = ["яблуко", "банан", "апельсин", "груша"]

print("Останній фрукт:", fruits[-1])         # груша
print("Передостанній:", fruits[-2])          # апельсин
````

````python
# Довжина списку - функція len()
fruits = ["яблуко", "банан", "апельсин", "груша"]

print("У списку", len(fruits), "елементів")
````

### Зміна списків — методи

Список можна змінювати: додавати, видаляти, змінювати елементи.

| Метод | Що робить |
|---|---|
| `append(x)` | Додати `x` в кінець списку |
| `insert(i, x)` | Вставити `x` на позицію `i` |
| `remove(x)` | Видалити перше входження `x` |
| `pop()` | Видалити і повернути останній елемент |
| `sort()` | Відсортувати список |
| `reverse()` | Перевернути список |

````python
# Додаємо елементи
shopping_list = ["хліб", "молоко"]
print("Початковий список:", shopping_list)

shopping_list.append("сир")
print("Після append:", shopping_list)

shopping_list.append("яйця")
print("Ще додали:", shopping_list)
````

````python
# Видаляємо елементи
shopping_list = ["хліб", "молоко", "сир", "яйця"]

shopping_list.remove("молоко")
print("Після видалення молока:", shopping_list)

# Змінюємо елемент за індексом
shopping_list[0] = "багет"
print("Після заміни хліба на багет:", shopping_list)
````

````python
# Сортування і реверс
numbers = [42, 7, 19, 3, 88, 25]
print("Початковий:", numbers)

numbers.sort()
print("Відсортований:", numbers)

numbers.reverse()
print("Перевернутий:", numbers)
````

# Частина 4. Словники (`dict`) — нова тема! 🆕

## Що таке словник?

**Словник** — це колекція пар `ключ : значення`. Уявіть справжній словник: ви шукаєте слово (це КЛЮЧ) — і знаходите його переклад або визначення (це ЗНАЧЕННЯ).

### Чим словник відрізняється від списку?

| Список | Словник |
|---|---|
| Доступ за номером (індексом) | Доступ за ключем (наприклад, ім'ям) |
| `students[0]` | `student["name"]` |
| Впорядкований | Неважливий порядок (доступ за ключем) |
| `[ ]` квадратні дужки | `{ }` фігурні дужки |

### Як виглядає словник

```python
student = {
    "name": "Олена",
    "age": 20,
    "city": "Київ"
}
```

Тут:
- `"name"`, `"age"`, `"city"` — це **ключі**
- `"Олена"`, `20`, `"Київ"` — це **значення**

### Коли використовувати словник?

Коли ви описуєте **один об'єкт з різними характеристиками** — як картка студента, профіль користувача, налаштування програми. Або коли треба зберегти **відповідності** — наприклад, ціни товарів, телефонна книга.

````python
# Створюємо словник зі студентом
student = {
    "name": "Олена Коваленко",
    "age": 20,
    "city": "Київ",
    "course": 3
}

# Виводимо весь словник
print(student)
````

````python
# Доступ до значень за ключем
student = {
    "name": "Олена Коваленко",
    "age": 20,
    "city": "Київ",
    "course": 3
}

# Звертаємось за ключем у квадратних дужках
print("Ім'я:", student["name"])
print("Місто:", student["city"])
print("Вік:", student["age"])
````

### Зміна словників

Словник, як і список, можна змінювати: додавати нові пари, змінювати наявні, видаляти.

````python
# Створюємо словник з цінами товарів
prices = {
    "хліб": 25,
    "молоко": 38,
    "сир": 120
}
print("Початкові ціни:", prices)

# Додаємо новий товар
prices["масло"] = 95
print("Додали масло:", prices)

# Змінюємо ціну
prices["хліб"] = 30
print("Хліб подорожчав:", prices)

# Видаляємо товар
del prices["сир"]
print("Сир закінчився:", prices)
````

````python
# Перевірка наявності ключа
prices = {"хліб": 30, "молоко": 38, "масло": 95}

if "хліб" in prices:
    print("Хліб коштує", prices["хліб"], "грн")

if "торт" in prices:
    print("Торт є")
else:
    print("Торта немає в нашому словнику")
````

### Корисні методи словників

| Метод | Що робить |
|---|---|
| `.keys()` | Повертає всі ключі |
| `.values()` | Повертає всі значення |
| `.items()` | Повертає пари (ключ, значення) |
| `.get(key)` | Повертає значення по ключу або `None`, якщо ключа немає |

````python
# Методи keys, values, items
prices = {"хліб": 30, "молоко": 38, "масло": 95, "сир": 120}

print("Усі ключі:", list(prices.keys()))
print("Усі значення:", list(prices.values()))
print("Усі пари:", list(prices.items()))
````

````python
# Метод get() - безпечний спосіб отримати значення
prices = {"хліб": 30, "молоко": 38}

# Звичайне звертання видасть помилку, якщо ключа немає:
# print(prices["торт"])   # ❌ KeyError

# А get() поверне None або вказане значення за замовчуванням:
print(prices.get("хліб"))                    # 30
print(prices.get("торт"))                    # None
print(prices.get("торт", "немає в продажу"))  # "немає в продажу"
````

### 🔥 Найважливіше: цикл `for` зі словником

Цикл `for` відмінно працює зі словниками. Є три способи перебору:

````python
# Спосіб 1: перебір КЛЮЧІВ (за замовчуванням)
prices = {"хліб": 30, "молоко": 38, "масло": 95, "сир": 120}

for key in prices:
    print(key)
````

````python
# Спосіб 2: перебір ЗНАЧЕНЬ
prices = {"хліб": 30, "молоко": 38, "масло": 95, "сир": 120}

for value in prices.values():
    print(value)
````

````python
# Спосіб 3: перебір ПАР ключ-значення (найкорисніший!)
prices = {"хліб": 30, "молоко": 38, "масло": 95, "сир": 120}

for product, price in prices.items():
    print(product, "коштує", price, "грн")
````

````python
# Практичний приклад: підрахунок загальної вартості покупок
prices = {
    "хліб": 30,
    "молоко": 38,
    "масло": 95,
    "сир": 120,
    "яйця": 65
}

total = 0
for product, price in prices.items():
    print(f"{product}: {price} грн")
    total = total + price

print("─" * 30)
print("Усього:", total, "грн")
````

````python
# Складніший приклад: рахуємо, скільки разів кожна літера зустрічається в слові
word = "програмування"
letter_count = {}   # порожній словник

for letter in word:
    if letter in letter_count:
        letter_count[letter] = letter_count[letter] + 1
    else:
        letter_count[letter] = 1

print(letter_count)

print("Підрахунок літер у слові", word + ":")
for letter, count in letter_count.items():
    print(f"  '{letter}' зустрічається {count} раз(ів)")
````

### Програма: Облік відвідуваності студентів

````python
# Список студентів
students = ["Олена", "Андрій", "Марія", "Богдан", "Софія"]

# Словник для зберігання відвідуваності
attendance = {}

print("=== Облік відвідуваності ===")
print(f"Сьогодні в групі {len(students)} студентів")
print()

# Перебираємо студентів і запитуємо, чи присутні
for student in students:
    answer = input(f"{student} присутній? (так/ні): ")
    if answer.lower() == "так":
        attendance[student] = "присутній"
    else:
        attendance[student] = "відсутній"

# Підрахунок результатів
present_count = 0
absent_count = 0

print()
print("=== Результати ===")
for student, status in attendance.items():
    print(f"{student}: {status}")
    if status == "присутній":
        present_count = present_count + 1
    else:
        absent_count = absent_count + 1

print()
print(f"Присутніх: {present_count}")
print(f"Відсутніх: {absent_count}")
print(f"Відвідуваність: {present_count / len(students) * 100:.0f}%")
````

````python
# Завдання 2. Телефонна книга
# Створіть словник з телефонами 5 знайомих (ім'я → номер). Напишіть програму, яка:

# Запитує у користувача ім'я
# Якщо ім'я є у словнику — виводить телефон
# Якщо немає — пропонує додати новий контакт

phone_numbers = {
    'Max': '+4918218218188',
    'Nastya': '+4900010101012',
    'Dmutro': '+4911100010102',
    'Jorik': '+4966677788999',
    'Danya': '+4911122211222'
}

name = input("Введіть ім'я: ")
while name != 'стоп':
    if name in phone_numbers.keys():
        print(f'Номер телефону {name}: {phone_numbers[name]}')
    else:
        phone = input("Такого контакту немає, введіть номер телефону: ")
        phone_numbers[name] = phone


print(phone_numbers)
````

````python
# Дано речення (рядок). Підрахуйте, скільки разів кожне СЛОВО зустрічається в реченні.
# Підказка: розбити рядок на слова можна методом .split().
# Наприклад: "привіт світ привіт".split() поверне ['привіт', 'світ', 'привіт'].

sentence = input("Введіть речення: ").lower().replace(",", "").replace(".", "").replace("!", "")
words = sentence.split()
word_count = {}

for word in words:
    if word not in word_count.keys():
        word_count[word] = 1
    else:
        word_count[word] = word_count[word] + 1

print(word_count)
````
