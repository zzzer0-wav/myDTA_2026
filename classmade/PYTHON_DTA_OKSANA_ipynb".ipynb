**Задача 1.5**  
Напишіть програму для перетворення секунд у дні, години, хвилини та секунди.

```
Вхід:   6785
Вихід:  0 day(s), 1 hour(s), 53 minute(s), 5 second(s).
```

````python
sec = int(input('Введіть кількість секунд: '))

days = sec // 86400
remaining_sec_after_days = sec % 86400

hours = remaining_sec_after_days // 3600
remaining_sec_after_hours = remaining_sec_after_days % 3600

mins = remaining_sec_after_hours // 60
final_secs = remaining_sec_after_hours % 60

print(f'{days} day(s), {hours} hour(s), {mins} minute(s), {final_secs} second(s).')
````

## 📖 Розгляд теоретичної частини

Для того, щоб повторити якийсь блок коду **`кілька разів`** або **`повторювати, доки виконується деяка умова`**, у Python реалізовані цикли. Вони є фундаментальною конструкцією, яка дозволяє повторювати виконання певного блоку коду кілька разів.



Існують два основних типи циклів: `for` та `while`


### Цикл `while`

Цикл while часто використовується, коли кількість ітерацій заздалегідь невідома або коли ітерації залежать від змінних, які можуть бути модифіковані в процесі виконання циклу.

Цикл while виконує блок коду, поки задана умова є істинною (True). Як тільки умова стає неправдивою (False), цикл закінчується.




```
while condition:
    # виконувати дії, поки condition є True
```



````python
while True :
  print('Hi'
````

````python
number = 5
while number == 5:
  print(number
````

````python
number = 5
while number  > 0 :
  number = number - 1
  print(number
````

````python
number = 5
while number  > 0 :
  number = number - 1
  if number % 2 == 0 :
    continue
  print(number)
````

````python
number = 5
while True :
  number = number - 1
  if number <= 0:
    break
  print(number)
````

### Цикл `for`

Цикл  `for`  перебирає елементи.

````python
for el in 1, 2, 3, 'hi':
  print(el)
````

Функція `range` у поєднанні з циклом `for` у Python є потужним інструментом для контролю повторення дій певну кількість разів. Цей механізм часто використовується для ітерації через послідовність чисел, що робить його особливо корисним для різних задач, від базових до складних алгоритмів.



Функція range створює послідовність чисел. Вона може бути використана різними способами:



`range(stop)`: Створює послідовність чисел від 0 до stop - 1.

`range(start, stop)`: Генерує числа від start до stop - 1.

`range(start, stop, step)`: Створює числа від start до stop - 1, з кроком step.

цикл `for` разом з `range` для ітерації через послідовність чисел виглядає так:

````python
range(5)
````

````python
for el in  range(5):
  print(el)
````

````python
for el in  range(2, 5):
  print(el)
````

````python
for el in  range(2, 15, 3):
  print(el)
````

# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

`dict()` - словник  
`d = {}` - створили порожній словник  
`d = {'key': 'value', 'key2': 'value2'}`
`key` - не змінні дані  
У одному словнику не може бути 2 однакових ключа!!  
`values` - можна змінювати, або взагалі задати порожніми.

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}

b.issubset(a) # True = b підмножина а
a.issuperset(b) # True = а надмножина b
a.isdisjoint(b) # False = перетин - чи незалежні дві множини (немає спільної частини)
````
