Типи результатів функцій

````python
my_dict = {
    1: 'one',
    2: 'two',
    3: 'three',
    4: 'four',
    5: 'five'
}
````

# NoneType нам показало результат (або ні) та це


````python
# Повертає NoneType, без зміни, виводить результат
def info(d):
    print(d)

# info(my_dict)
type(info(my_dict))
````

````python
# Повертає NoneType, змінює початкові дані, виводить результат
def add_el_viev(d):
    d[6] = 'six'
    print(d)

# add_el_viev(my_dict)
type(add_el_viev(my_dict))
````

````python
# Повертає NoneType, змінює початкові дані, не виводить результат
def add_el(d):
    d[6] = 'six'

print(f'my dict: {my_dict}')
type(info(my_dict))
````

````python
print(type(add_el(my_dict)))
display(type(add_el(my_dict)))
````

Функція повертає результат для обробки

````python
def sum_num(a, b):
    print(a + b)

type(sum_num(5, 7))
````

````python
def sum_num(a, b):
    return a + b

print(sum_num(5, 7))
print(type(sum_num(5, 7)))

print()

sum_num(5, 7) + 10
````

````python
my_dict = {
    1: 'one',
    2: 'two',
    3: 'three',
    4: 'four',
    5: 'five'
}
````

````python
def dict_value(d, key):
    print(d[key])

dict_value(my_dict, 3)
````

````python
def dict_value(d: dict, key: int|str) -> str:
    return d[key] # my_dict[3]

dict_value(my_dict, 3) + '333'
````

````python
def new_func(d: dict, key: int|str) -> str:
    return d[key]

new_func('qwerty', 3)
````

````python
def func_3(a=5, b=7):
    print(a + b)

func_3()  # 5 + 7
func_3(10) # 10 + 7
func_3(10, 10)  # 10 + 20 (a=10, b=20)
func_3(b=10)  # 5 + 10 (b=10)
````

````python
def func_4(a, b=7):
    print(a + b)

func_4(10)  # 10 + 7 (a=10)
func_4(10, 20)  # 10 + 20 (a=10, b=20)
````
